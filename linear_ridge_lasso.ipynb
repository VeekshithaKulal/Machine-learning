```python
import pandas as pd

# To read csv file
df=pd.read_csv("/content/car_purchasing 2.csv",encoding="latin1")
df
```

```python
# To copy
df1=df.copy()
df1
```

```python
# To retrieve the information of dataset
df1.info()
```

```python
# To get the summary statistics
df1.describe()
```

```python
# To drop unnecessay columns
df1.drop(['customer name','customer e-mail','country','gender'],
         axis=1, inplace=True, errors='ignore')

```

```python
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df1.corr(), annot=True, cmap="rocket")
plt.show()
```

```python
# define x and y
# x contains all the rows and all the columns expect last column
# y contains all rows and last column
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
```

```python
# converting into 2 dimensional array
y=y.values.reshape(-1,1)
y
```

```python
# Feature scaling(To compare one feature to another)
from sklearn.preprocessing import MinMaxScaler
# converts values from 0 to 1
scaler = MinMaxScaler()
x= scaler.fit_transform(x)
y=scaler.fit_transform(y)
x
```

```python
y
```

```python
# 80% data for training and 20% data for testing
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,shuffle=True, random_state=42)
# parameter shuffle used to shuffle old records and recent records
# parameter random state is used for ensuring same split every time
x_train, x_test, y_train, y_test
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression

# to train model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# how much model best fits the training data
model.score(x_train,y_train)
```

```python
# predict numeric values
y_pred=model.predict(x_test)
y_pred
```

```python
# to calculate r2 score and mean square error
from sklearn.metrics import mean_squared_error,r2_score
# r2 shoiuld be more
print("R² Score:", r2_score(y_test, y_pred))
# mse should be less
print("MSE:", mean_squared_error(y_test, y_pred))
```

# **Ridge Regression**

```python
from sklearn.linear_model import Ridge
model2=Ridge()
model2.fit(x_train,y_train)
model2.score(x_train,y_train)
y_pred2=model2.predict(x_test)
print("R² Score:", r2_score(y_test, y_pred2))
print("MSE:", mean_squared_error(y_test, y_pred2))
```

# **Lasso regression**

```python
from sklearn.linear_model import Lasso
model3=Lasso()
model3.fit(x_train,y_train)
model3.score(x_train,y_train)
y_pred3=model3.predict(x_test)
print("R² Score:", r2_score(y_test, y_pred3))
print("MSE:", mean_squared_error(y_test, y_pred3))
```

# **Plotting Graph**

```python
import matplotlib.pyplot as plt

plt.plot(y_test, y_pred, label="Linear Regression",color="pink",marker="o")
plt.plot(y_test, y_pred2, label="Ridge Regression",color="yellow",marker="o")
plt.plot(y_test, y_pred3, label="Lasso Regression",color="purple",marker="o")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear vs Ridge vs Lasso Regression")
plt.legend()
plt.show()
```
